```python
import pandas as pd
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
print(boston.DESCR)
```

```python
boston.feature_names
```

```python
X_orig = pd.DataFrame(boston.data,columns=boston.feature_names)
X_orig.head()
```

```python
X_orig.info()
```

```python
y_orig = pd.Series(boston.target,name='MEDV')
y_orig[0:6]
```

 ### Crear Datos de entrenamiento y prueba: Train data and test data
 
 

```python
# importacion de librerias 
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# obtenemos los set de training y de tes desde la data original con un ratio de 75% - 25% "by default"

from sklearn.model_selection import train_test_split

X, X_test, y, y_test = train_test_split(X_orig,y_orig,random_state=1)

# Donde X=set de training, X_test = set de prueba (test), y = set de training para x, y_test= set de test para Y

```

```python
df = pd.concat([X,y], axis=1)
```

```python
df.info()
```

```python
# Verificando valores nulos

X.isna().sum()
```

```python
y.isna().sum()
```
