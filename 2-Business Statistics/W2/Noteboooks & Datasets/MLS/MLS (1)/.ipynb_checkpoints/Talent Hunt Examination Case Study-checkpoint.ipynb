### Background:
A research institute conducts a  talent hunt examination every year to hire people who can work on various research projects in the field of Mathematics and Computer Science. ‘A2Z’ institute provides a preparatory program to help the aspirants prepare for the talent hunt exam. The institute has a good record of helping many students clear the exam. Before the application for the next batch starts, the institute wants to attract more aspirants to their program. For this, the institute wants to assure the aspiring students of the quality of results obtained by students enrolled in their program in recent years. 

However, one challenge in estimating an average score is that every year the exam’s difficulty level varies a little, and the distribution of scores also changes accordingly. The institute keeps a track of the final scores of its alumni who attempted the exam previously. A dataset constituted of a simple random sample of final scores of 600 aspirants from the last three years is prepared by the institute.

### Objective:

The institute wants to provide an estimate of the average score obtained by aspirants enrolled in their program. Keeping in mind the variation in scores every year, the institute wants to estimate a range of scores instead of a single estimate of the average score, to make the estimate more robust. It is known from the history of scores in the exam that the standard deviation of the scores is 10. And the cut-off score in the most recent year was 84.

Using the data of the final scores, the institute wants to perform the following task to complete the objective:

1. Construct a 95% confidence interval for the mean score that an aspirant has scored after enrolling in the institute’s program.


2. Perform a hypothesis test (taking alpha = 5%) to check if the mean score of the aspirants enrolled in the program is less than the last year’s cut-off score, which is 84. Using the test results, verify the confidence interval constructed for the mean score.


### Dataset: 
The dataset contains the final scores of 600 aspirants enrolled in the institute’s program in the last three years.

### 1. Construct a 95% confidence interval for the mean score secured by an aspirant after enrolling in the institute’s program.

Before calculating a confidence interval for the scores' mean, let's check a few assumptions:
- We have a random sample according to the institute
- The size of the sample  $n$ is > 30 so the CLT applies and the mean follows a normal distribution
- The standard deviation is known and equal to 10 so the Z-score follows a normal distribution
Hence, we can derive the confidence interval using the normal distribution

**Note**: if the standard deviation was unknown, the Z-score follows a student distribution and the confidence interval for the mean should be derived from a student distribution. However, given the size of our sample, this makes no difference

```python
# import the required function
from scipy.stats import norm, t

# calculate the value of sample mean. Population standard deviation is known to be 10
x_bar = df.final_score.mean()
sigma = 10

n = 600 #sample size

alpha = 0.05

# construct the confidence interval
np.round(norm.interval(1 - alpha, loc=x_bar, scale=sigma / np.sqrt(n)), 2)
```

The 95% confidence interval obtained for the mean score is [84.02, 85.62]. This means that it is 95% likely that an aspirant enrolled in the institute's program will score, on average, between 84.02 and 85.62.

**Question**: Can we use this interval to test the hypotheses of question 2 ? (Answer below)

#### Adel add-on: effect of size and the knowledge of the standard deviation

- Whenever the CVLT assumptions are fullfilled, the mean always follows a normal distribution (regardless of whether we know the standard deviation of the variables)
- To calculate the confidence interval:
    - *Standard deviation is known*: use normal distribution interval with the known std
    - *Standard deviation is not knwon and size is large*: estimate the std and then use the normal dist intervals 
    - *Standard deviation is not known and size is small*: estimate the std and use t-distribution intervals

- When the sample size is smaller, the confidence interval of the mean is larger

```python
np.round(norm.interval(0.95, loc=x_bar, scale=sigma / np.sqrt(100)), 2)
```

- When the sample size is smaller and the standard deviation unknown, the t-distribution provides largers confidence intervals

```python
n_small = 35
```

```python
np.round(norm.interval(0.95, loc=x_bar, scale=df.final_score.std() / np.sqrt(n_small)), 2)
```

```python
np.round(t.interval(0.95, df = n_small, loc=x_bar, scale=df.final_score.std() / np.sqrt(n_small)), 2)
```

#### Adel add-on: interval formula for the centered confidence interval

- Notice that the interval in scipy function is calculated using the following formula
$$\left [\bar x + \frac{\sigma}{\sqrt{n}} \times z_{\alpha/2}, \bar x + \frac{\sigma}{\sqrt{n}} \times z_{1-\alpha/2} \right]$$
where $z_{\alpha}$ is the $\alpha-$quantile of the standard normal distribution. Let's check it

```python
np.round((x_bar + sigma*norm.ppf(alpha/2)/np.sqrt(n),
         x_bar + sigma*norm.ppf(1 - alpha/2)/np.sqrt(n)), 2)
```

### 2. Perform a hypothesis test (taking alpha = 5%) to check if the mean score of the aspirants enrolled in the program is less than last year’s cut-off score of 84. Using the test results, verify the confidence interval constructed for the mean score.

**Let's write the null and alternative hypotheses**

Let $\mu$ be the mean final_score of all aspirants who enroll in the institute's program.

The null and alternative hypothesis can be written as:

>$H_0: \mu >= 84$

against the alternate hypothesis

> $H_a: \mu < 84$

**Are the assumptions of Z-test satisfied?**

*   Samples are drawn from a normal distribution - Since the sample size is 600(which is > 30), Central Limit Theorem states that the distribution of sample means will be approximately normal.  
*   Observations are from a simple random sample - the institute collected a simple random sample
*   Standard deviation is known - Yes


Voila! We can use Z-test for this problem.

**Calculate the Z-statistic value**

$$Z_{score} = \frac{\bar x - \mu}{\sigma_{mean}}, ~~~~~ \sigma_{mean} = \frac{\sigma}{\sqrt{n}}$$

The Z test statistic follows a standard normal distribution.

```python
# set the values of population mean and population standard deviation 
mu, sigma = 84, 10

# calculate the value of the sample mean from the data
x_bar = df.final_score.mean()


# calculate the test statistic
test_stat = (x_bar - mu) / (sigma/np.sqrt(600))
test_stat
```

**Calculating the p-value**

The p-value reflects how likely is the observed quantity *given the assumption of the null hypothesis*. Hence, its calculation depends on the type of test (one-tailed, two-tailed) used.

Follow these rules to decide how to calculate the p-value:

- *For a left-tailed test* ($H_a : \mu < 84)$: mean values $\bar x$ much smaller than 84 (Eg: $\bar x = 0$) are unlikely if the null hypothesis holds, i.e. if $\mu >= 84$. Hence, the p-value is the probability of getting **at most** the observed mean value. The area to consider is to the left and 

$$p-value = CDF(\mbox{test_statistic})$$
<br />

- *For a right-tailed test* ($H_a : \mu > 84)$: mean values $\bar x$ much bigger than 84 (Eg: $\bar x = 200$) are unlikely if the null hypothesis holds, i.e. if $\mu <= 84$. Hence, the p-value is the probability of getting **at least** the observed mean value. The area to consider is to the right and 

$$p-value = 1 - CDF(\mbox{test_statistic})$$
<br />

- *For a two-tailed test* ($H_a : \mu \neq 84)$: mean values $\bar x$ much bigger or smaller than 84 are unlikely if the null hypothesis holds, i.e. if $\mu = 84$. Hence, the p-value is the probability of getting at least the absolute value of the observed mean value. The areas to consider is to the left of $-|\mbox{test_statistic}|$ and to the right of $|\mbox{test_statistic}|$ and 

$$p-value = CDF(-|\mbox{test_statistic}|) + 1 - CDF(|\mbox{test_statistic}|)$$





Since our test is left-tailed, the p-value is the area to the left of the test statistics as shown below

```python
# import the required function
from scipy.stats import norm

# plotting the distribution of Z test statistic along with the computed test statistic
# We are plotting the distributions here to better visualize the calculations
x = np.linspace(-4, 4, 100) # create an array of 100 numbers starting from -4 and ending at 4
plt.plot(x, norm.pdf(x, 0, 1)) # plot the pdf of the normal distribution
plt.axvline(x = test_stat, c = 'b') # draw a vertical red line through the mentioned point
## Adel add-on for the area below the curve
x1 = np.linspace(-4, test_stat, 50)
plt.fill_between(x1, norm.pdf(x1, 0, 1), color='b') # fill the area under the curve before the critical value with red color
plt.annotate('p-value', (-3, 0.20), color = 'b') # annotate the mentioned text at the mentioned location
plt.annotate('Test statistic', (2.2, 0.35), color = 'blue')
plt.show() # display the plot
```

```python
# calculate the p-value using the cdf() function
norm.cdf(test_stat)
```

### Insight

- Rules:
    - When the p-value is bigger than the significance level $\alpha$: I fail to reject $H_0$
    - When the p-value is smaller than the significance level $\alpha$: I reject $H_0$

As the p-value ~0.98 is greater than the level of significance (0.05), we fail to reject the null hypothesis. Thus, we do not have enough statistical evidence to say that the mean final score of all aspirants who enroll in the institute's program is less than last year's cutoff of 84.

**NOTE**: We can **NOT** use the centered confidence interval calculated in Q1 to test a one-sided hypothesis (Check with 0.982*X_bar instead of X_bar: the hypothesis will be rejected although the value 84 is in the centered confidence interval). 

We need to construct a one-sided interval with the folllowing formula
$$\left [ -\infty, \bar x + \frac{\sigma}{\sqrt{n}} \times z_{1-\alpha} \right ]$$
Hence the one-sided confidence interval

```python
np.round((-np.inf, x_bar+ sigma * norm.ppf(0.95)/np.sqrt(n)), 2)
```

Since the value 84 is in that interval, the null hypothesis is not rejected.

#### Adel add-on : critical value approach

Let's calculate the critical value for the test. The critical value $S_{crit}$ is such that 
$$P(Z < S_{crit}) = 0.05$$

Since $Z$ follows a standard normal distribution, the critical value is

```python
norm.ppf(0.05)
```

Since the test statistic is bigger than the critical value, the null hypothesis is not rejected

- we could also calculate the critical value directly in terms of scores from the distribution mean

```python
critical_value = np.round(norm.ppf(0.05, loc = mu, scale = sigma/np.sqrt(n)), 2)
critical_value
```

```python
# plotting the distribution of Z test statistic along with the computed test statistic
# We are plotting the distributions here to better visualize the calculations
x = np.linspace(82, 86, 100) # create an array of 100 numbers starting from -4 and ending at 4
plt.plot(x, norm.pdf(x, mu, sigma/np.sqrt(n))) # plot the pdf of the normal distribution
#plt.ylim((0, 0.050))
plt.axvline(x = x_bar, c = 'b') # draw a vertical red line through the mentioned point
plt.axvline(x = critical_value, c = 'r')
x1 = np.linspace(82, critical_value, 50)
plt.fill_between(x1, norm.pdf(x1, mu, sigma/np.sqrt(n)), color='r') # fill the area under the curve before the critical value with red color
plt.annotate('Reject Null', (82.4, 0.155), color = 'r') # annotate the mentioned text at the mentioned location
plt.annotate('Observed mean', (85, 0.45), color = 'blue')
plt.show() # display the plot
```

Since the observed mean is larger than the critical value, the null hypothesis is not rejected. You could try an observed mean X_bar lower than 83.33 and the p-value approach and you will find that the p-value is smaller than 0.05
