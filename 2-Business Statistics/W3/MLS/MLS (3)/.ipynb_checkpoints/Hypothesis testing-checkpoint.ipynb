```python
import numpy as np
from scipy.stats import t, norm
import matplotlib.pyplot as plt
```

```python
plt.rcParams['text.usetex'] = True
```

```python
alpha = 0.05
```

## Rejection zone

```python
critical_value = norm.ppf(1-alpha/2)
```

```python
x = np.linspace(-4, 4, 100) # create an array of 100 numbers starting from -4 and ending at 4
plt.plot(x, norm.pdf(x, 0, 1)) # plot the pdf of the normal distribution
#plt.axvline(x = test_stat, c = 'b') # draw a vertical red line through the mentioned point
## Adel add-on for the area below the curve
x1 = np.linspace(4, critical_value, 50)
plt.axvline(x = critical_value, c = 'r')
plt.axvline(x = -critical_value, c = 'r')
plt.fill_between(x1, norm.pdf(x1, 0, 1), color='r') # fill the area under the curve before the critical value with red color
plt.fill_between(-x1, norm.pdf(x1, 0, 1), color='r')
plt.annotate('Rejection region', (-1., 0.12), color = 'r', fontsize=14)
plt.annotate('$\\mbox{Area} ~~\\alpha/2$', (2.5, 0.03), color = 'r', fontsize=14)
plt.annotate('$\\mbox{Area} ~~\\alpha/2$', (-4, 0.03), color = 'r', fontsize=14)
plt.arrow(0, 0.1, 1.5, -0.05, color = 'r', shape = 'full', head_width = 0.02)
plt.arrow(0, 0.1, -1.5, -0.05, color = 'r', shape = 'full', head_width = 0.02)
plt.annotate('Critical value (+)', (2.3, 0.30), color = 'r', fontsize=14) # annotate the mentioned text at the mentioned location
plt.annotate('Critical value (-)', (-4.3, 0.30), color = 'r', fontsize=14)
plt.title('$H_a: \\mu \\neq \\mu_0$ ', fontsize = 20)
plt.axis('off')
#plt.annotate('Test statistic', (2.2, 0.35), color = 'blue')
plt.tight_layout()
plt.show() # display the plot
```

## P-value approach

```python
testStat = 2
criticalVal = norm.ppf(alpha)
```

```python
x = np.linspace(-4, 4, 100) # create an array of 100 numbers starting from -4 and ending at 4
plt.plot(x, norm.pdf(x, 0, 1)) # plot the pdf of the normal distribution
#plt.axvline(x = test_stat, c = 'b') # draw a vertical red line through the mentioned point
## Adel add-on for the area below the curve
x1 = np.linspace(-4, testStat, 50)
xcrit = np.linspace(-4, criticalVal, 50)
plt.axvline(x = testStat, c = 'b')
plt.axvline(x = criticalVal, c = 'r')
plt.fill_between(x1, norm.pdf(x1, 0, 1), color='b', alpha = 0.2) # fill the area under the curve before the critical value with red color
plt.fill_between(xcrit, norm.pdf(xcrit, 0, 1), color='r', alpha = 0.2)
#plt.fill_between(-x1, norm.pdf(x1, 0, 1), color='b')
#plt.annotate('Rejection region', (-1., 0.12), color = 'r', fontsize=14)
#plt.annotate('$\\mbox{Area} \\newline ~~p-value$', (2.5, 0.07), color = 'b', fontsize=14)
#plt.annotate('$\\mbox{Area} \\newline ~~p-value$', (-4, 0.07), color = 'b', fontsize=14)
#plt.arrow(0, 0.1, 1.5, -0.05, color = 'r', shape = 'full', head_width = 0.02)
#plt.arrow(0, 0.1, -1.5, -0.05, color = 'r', shape = 'full', head_width = 0.02)
plt.annotate('$Z$', (2.1, 0.30), color = 'b', fontsize=14) # annotate the mentioned text at the mentioned location
#lt.annotate('$Z$', (-2., 0.30), color = 'b', fontsize=14)
plt.annotate('Critical value', (-3.3, 0.30), color = 'r', fontsize=14)
plt.annotate('Rejection region', (-4.1, 0.05), color = 'r', fontsize=14)
plt.annotate('p-value', (0, 0.05), color = 'b', fontsize=14)
plt.title('$H_a: \\mu < 0$ ', fontsize = 20)
#plt.axis('off')
#plt.annotate('Test statistic', (2.2, 0.35), color = 'blue')
plt.tight_layout()
plt.show() # display the plot
```
